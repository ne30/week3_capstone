<h2> Importing request, bs4, numpy and pandas </h2>

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

<h2>Url using to scrap the html code </h2>
<br>

```python
url = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
```

```python
soup = BeautifulSoup(url,'lxml')
print(soup.prettify)
```

```python
table = soup.find('table',{'class':'wikitable sortable'})
```

```python
table
```

```python
table_row = table.find_all('tr')
```

```python
table_row
```

<h1> scraped data to dataframe</h1>
<br>

```python
l = []
for tr in table_row:
    td = tr.find_all('td')
    row = [tr.text.strip() for tr in td]
    l.append(row)
df = pd.DataFrame(l, columns=["A", "B", ...])
```

```python
df.head()
```

```python
df.shape[0]
```

```python
df.columns = ['PostalCode', 'Borough', 'Neighborhood']
```

```python
for i in range(df.shape[0]):
    if ((df.iloc[i][1]=="Not assigned")or(df.iloc[i][2]=="Not assigned")):
        df.iloc[i][0] = np.nan
        
```

```python
df.dropna(inplace = True)
```

```python
temp = ""
for i in range(df.shape[0]):
    if temp != df.iloc[i][0]:
        temp = df.iloc[i][0]
        continue
    else:
        df.iloc[i][2] = df.iloc[i][2] + ',' + df.iloc[i-1][2]
        df.iloc[i-1][2] = np.nan
```

<h1> Visualisation </h1>

```python
import folium

map_toronto = folium.Map(location=[df.iloc[0][3], df.iloc[0][4]], zoom_start=10)
for lat, lng, borough, neighborhood in zip(df["Latitude"], df["Longitude"], df["Borough"], df["Neighborhood"]):
     label = '{}, {}'.format(neighborhood, borough)
     label = folium.Popup(label, parse_html=True)
     folium.CircleMarker(
     [lat, lng],
     radius=5,
     popup=label,
     color='blue',
     fill=True,
     fill_color='#3186cc',
     fill_opacity = 0.7,
     parse_html=False).add_to(map_toronto)
map_toronto
```
